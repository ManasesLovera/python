## Python syntax

```python
# This is a comment
"""
This is a multi-line comment
"""

# Case sensitivity
a = 5
A = 10
print(a)  # prints 5
print(A)  # prints 10

# Multiple expressions on a single line
x = 5; y = 10; z = x + y; 

# Line continuation
x = 5 + \
    10 + \
    15

# Variable assignment
x = 5
y = 10
z = x + y
print(z)  # prints 15

# Data types
integer_var = 42
float_var = 3.14
string_var = "Hello, World!"
boolean_var = True
list_var = [1, 2, 3, 4, 5]
tuple_var = (1, 2, 3)
dict_var = {"key": "value"}
set_var = {1, 2, 3}

# Type conversion
x = 5
y = float(x)  # converts integer to float
z = str(x)  # converts integer to string
print(y)  # prints 5.0
print(z)  # prints "5"

# Type checking
print(type(x))  # prints <class 'int'>
print(type(y))  # prints <class 'float'>
print(type(z))  # prints <class 'str'>
print(type(list_var))  # prints <class 'list'>
print(type(tuple_var))  # prints <class 'tuple'>
print(type(dict_var))  # prints <class 'dict'>
print(type(set_var))  # prints <class 'set'>
print(type(boolean_var))  # prints <class 'bool'>

# Arithmetic operations
a = 10
b = 3
addition = a + b
subtraction = a - b
multiplication = a * b
division = a / b
modulus = a % b
exponentiation = a ** b
floor_division = a // b
print(addition)  # prints 13
print(subtraction)  # prints 7
print(multiplication)  # prints 30
print(division)  # prints 3.3333333333333335
print(modulus)  # prints 1
print(exponentiation)  # prints 1000
print(floor_division)  # prints 3

# Comparison operations
x = 5
y = 10
print(x == y)  # prints False
print(x != y)  # prints True
print(x < y)  # prints True
print(x > y)  # prints False
print(x <= y)  # prints True
print(x >= y)  # prints False

# Logical operations
x = True
y = False
print(x and y)  # prints False
print(x or y)  # prints True
print(not x)  # prints False

# Bitwise operations
x = 5  # 0101 in binary
y = 3  # 0011 in binary
print(x & y)  # prints 1 (0001 in binary)
print(x | y)  # prints 7 (0111 in binary)
print(x ^ y)  # prints 6 (0110 in binary)
print(~x)  # prints -6 (inverts bits)
print(x << 1)  # prints 10 (1010 in binary)
print(x >> 1)  # prints 2 (0010 in binary)

# if statement
if x < y:
    print("x is less than y")
else:
    print("x is greater than or equal to y")

# for loop
for i in range(5):
    print(i)  # prints numbers from 0 to 4

# for loop with list
for item in list_var:
    print(item)  # prints each item in the list

# while loop
while x < 10:
    print(x)
    x += 1  # increments x by 1

# Functions
def add(a, b):
    return a + b

# Input and Output
name = input("Enter your name: ")
print("Hello, " + name)

# Exception handling
try:
    result = 10 / 0
except ZeroDivisionError:
    print("You can't divide by zero!")

# Classes and Objects
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        print("Woof! My name is " + self.name)

dog = Dog("Buddy")
dog.bark()  # prints "Woof! My name is Buddy"

# List comprehensions
squared_numbers = [x**2 for x in range(10)]
print(squared_numbers)  # prints squares of numbers from 0 to 9

# Lambda functions
add_lambda = lambda a, b: a + b
print(add_lambda(5, 10))  # prints 15

# Map, Filter, Reduce
from functools import reduce
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers)  # prints squares of numbers from 1 to 5
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # prints even numbers from 1 to 5
sum_of_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_of_numbers)  # prints sum of numbers from 1 to 5

# File I/O
# Writing to a file
with open("example.txt", "w") as file:
    file.write("Hello, World!")
# Reading from a file
with open("example.txt", "r") as file:
    content = file.read()
    print(content)  # prints "Hello, World!"

# Regular expressions
import re
pattern = r"\d+"
text = "There are 123 apples"
matches = re.findall(pattern, text)
print(matches)  # prints ['123']

# JSON handling
import json
data = {
    "name": "John",
    "age": 30,
    "city": "New York"
}
json_string = json.dumps(data)
print(json_string)  # prints JSON string
parsed_data = json.loads(json_string)
print(parsed_data)  # prints parsed JSON data

# Date and Time
from datetime import datetime
now = datetime.now()
print(now)  # prints current date and time
formatted_date = now.strftime("%Y-%m-%d %H:%M:%S")
print(formatted_date)  # prints formatted date and time

# Time delay
import time
print("Waiting for 2 seconds...")
time.sleep(2)  # waits for 2 seconds
print("Done!")

# List methods
my_list = [1, 2, 3, 4, 5]
my_list.append(6)  # adds 6 to the end of the list
print(my_list)  # prints [1, 2, 3, 4, 5, 6]
my_list.remove(3)  # removes 3 from the list
print(my_list)  # prints [1, 2, 4, 5, 6]
my_list.sort()  # sorts the list
print(my_list)  # prints sorted list
my_list.reverse()  # reverses the list
print(my_list)  # prints reversed list
my_list.clear()  # clears the list
print(my_list)  # prints []

# Dictionary methods
my_dict = {"name": "John", "age": 30}
my_dict["city"] = "New York"  # adds a new key-value pair
print(my_dict)  # prints {'name': 'John', 'age': 30, 'city': 'New York'}
my_dict.pop("age")  # removes the key 'age'
print(my_dict)  # prints {'name': 'John', 'city': 'New York'}
my_dict.clear()  # clears the dictionary
print(my_dict)  # prints {}

# Set methods
my_set = {1, 2, 3}
my_set.add(4)  # adds 4 to the set
print(my_set)  # prints {1, 2, 3, 4}
my_set.remove(2)  # removes 2 from the set
print(my_set)  # prints {1, 3, 4}
my_set.clear()  # clears the set
print(my_set)  # prints set()

# Tuple methods
my_tuple = (1, 2, 3)
print(my_tuple)  # prints (1, 2, 3)
# Tuples are immutable, so we cannot change them
# but we can convert them to a list, modify, and convert back
my_list = list(my_tuple)  # converts tuple to list
my_list.append(4)  # adds 4 to the list
my_tuple = tuple(my_list)  # converts list back to tuple
print(my_tuple)  # prints (1, 2, 3, 4)

# String methods
my_string = "Hello, World!"
print(my_string.lower())  # prints "hello, world!"
print(my_string.upper())  # prints "HELLO, WORLD!"
print(my_string.replace("World", "Python"))  # prints "Hello, Python!"
print(my_string.split(", "))  # prints ['Hello', 'World!']
print(my_string.find("World"))  # prints 7
print(my_string.count("o"))  # prints 2
print(my_string.startswith("Hello"))  # prints True
print(my_string.endswith("!"))  # prints True
print(my_string.isalpha())  # prints False
print(my_string.isdigit())  # prints False
print(my_string.isalnum())  # prints False
print(my_string.islower())  # prints False
print(my_string.isupper())  # prints False
print(my_string.isspace())  # prints False
print(my_string.strip())  # prints "Hello, World!"
print(my_string.lstrip())  # prints "Hello, World!"
print(my_string.rstrip())  # prints "Hello, World!"
print(my_string.title())  # prints "Hello, World!"
print(my_string.capitalize())  # prints "Hello, world!"
print(my_string.zfill(20))  # prints "0000000000Hello, World!"
print(my_string.center(20))  # prints "   Hello, World!   "
print(my_string.ljust(20))  # prints "Hello, World!       "
print(my_string.rjust(20))  # prints "       Hello, World!"
print(my_string.splitlines())  # prints ['Hello, World!']
print(my_string.split(" "))  # prints ['Hello,', 'World!']
print(my_string.join(["Hello", "World"]))  # prints "HelloHello, World!World"
print(my_string.index("World"))  # prints 7
print(my_string.rindex("o"))  # prints 8
print(my_string.rfind("o"))  # prints 8
print(my_string.partition("World"))  # prints ('Hello, ', 'World', '!')
print(my_string.rpartition("o"))  # prints ('Hello, W', 'o', 'rld!')
print(my_string.split("o"))  # prints ['Hell', ', W', 'rld!']
print(my_string.rsplit("o"))  # prints ['Hell', ', W', 'rld!']
print(my_string.split(" ", 1))  # prints ['Hello,', 'World!']
print(my_string.rsplit(" ", 1))  # prints ['Hello,', 'World!']
print(my_string.split(" ", 2))  # prints ['Hello,', 'World!']
print(my_string.rsplit(" ", 2))  # prints ['Hello,', 'World!']
print(my_string.split(" ", 0))  # prints ['Hello,', 'World!']
print(my_string.rsplit(" ", 0))  # prints ['Hello,', 'World!']
print(my_string.split(" ", -1))  # prints ['Hello,', 'World!']
print(my_string.rsplit(" ", -1))  # prints ['Hello,', 'World!']
print(my_string.split(" ", 1))  # prints ['Hello,', 'World!']
print(my_string.rsplit(" ", 1))  # prints ['Hello,', 'World!']
print(my_string.split(" ", 2))  # prints ['Hello,', 'World!']
print(my_string.rsplit(" ", 2))  # prints ['Hello,', 'World!']
print(my_string.split(" ", 0))  # prints ['Hello,', 'World!']
print(my_string.rsplit(" ", 0))  # prints ['Hello,', 'World!']
print(my_string.split(" ", -1))  # prints ['Hello,', 'World!']
```
