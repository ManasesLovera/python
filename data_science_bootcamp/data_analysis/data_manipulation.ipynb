## Data Manipulation and Analysis with Pandas
Data manipulation and analysis are key tasks in any data science or data analysis project. Pandas provides a wide range or functions for data manipulation and analysis, making it easier to clean, transform, and extract insights from data. In this lesson, we will cover various data manipulation and analysis techniques using Pandas.

```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df.tail(3)  # Display the last 3 rows of the DataFrame
```

```python
df.describe()  # Get a summary of the DataFrame
```

```python
df.dtypes
```

```python
# Handling missing values
df.isnull().any()
```

```python
num = df['Value'].sum()  # Sum of the 'Value' column
df = df.fillna(0)  # Fill missing values with 0
print(f"All values:\n{df['Value'].tolist()}")  # Print all values in the 'Value' column

print(f"Total Value: {num}")
```

```python
df.drop(columns=['Sales_fillNa'], inplace=True)  # Drop the 'Sales_fillNa' column
```

```python
# Filling missing values with the mean of the column
df['Value_fillNa'] = df['Value'].fillna(df['Value'].mean())
df
```

```python
# Rename columns
df.rename(columns={'Value_fillNa': 'Value Filled'}, inplace=True)  # Rename the column
df.rename(columns={'Date': 'Sales Date'}, inplace=True)  # Rename the 'Date' column
df.head() # Display the first few rows of the DataFrame
```

```python
df['Value + 10%'] = df['Value'].apply(lambda x: x * 1.1)
```

```python
import numpy as np

"""
Quiero saber si hay una relacion entre la cantidad de amigos y las horas que se pasan al da en redes sociales. el indice representa
cada persona, el array num_friends representa la cantidad de amigos que tiene cada persona, y el array daily_hours representa las horas que pasan al da en redes sociales.
"""
num_friends = np.array([100, 150, 200, 250, 300])
daily_hours = np.array([1, 2, 3, 4, 5])

import matplotlib.pyplot as plt
plt.scatter(num_friends, daily_hours)
plt.title('Relationship between Number of Friends and Daily Hours on Social Media')
plt.xlabel('Number of Friends')
plt.ylabel('Daily Hours on Social Media')
plt.show()  # Display the scatter plot

# Covariance and correlation with numpy
covariance = np.cov(num_friends, daily_hours)[0][1]  # Calculate covariance
correlation = np.corrcoef(num_friends, daily_hours)[0][1]  # Calculate correlation
print(f"Covariance: {covariance}")  # Print covariance
print(f"Correlation: {correlation}")  # Print correlation
# Save the modified DataFrame to a new CSV file
#df.to_csv('data_modified.csv', index=False)  # Save the DataFrame without the
```
